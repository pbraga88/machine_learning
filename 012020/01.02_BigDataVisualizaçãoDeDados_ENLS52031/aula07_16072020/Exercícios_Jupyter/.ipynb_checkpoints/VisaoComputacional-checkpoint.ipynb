```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
```

```python
imgCrateraLua = cv2.imread('Imagens/CrateraLua.jpg')
imgSistemaSolar = cv2.imread('Imagens/SistemaSolar.jpg')
imgBuracoNegro = cv2.imread('Imagens/BuracoNegro.jpg')
imgOnibusEspacial = cv2.imread('Imagens/OnibusEspacial.jpg')

imagemCrateraLua = mpimg.imread('Imagens/CrateraLua.jpg')
imagemSistemaSolar = mpimg.imread('Imagens/SistemaSolar.jpg')
imagemBuracoNegro = mpimg.imread('Imagens/BuracoNegro.jpg')
imagemOnibusEspacial = mpimg.imread('Imagens/OnibusEspacial.jpg')
```

```python
histCrateraLua = cv2.calcHist([imgCrateraLua],[0],None,[256],[0,256]) 
plt.plot(histCrateraLua) 
plt.show() 
crateraLua = plt.imshow(imagemCrateraLua)
```

```python
histSistemaSolar = cv2.calcHist([imgSistemaSolar],[0],None,[256],[0,256]) 
plt.plot(histSistemaSolar) 
plt.show() 

plt.imshow(imagemSistemaSolar)
```

```python

histBuracoNegro = cv2.calcHist([imgBuracoNegro],[0],None,[256],[0,256]) 
plt.plot(histBuracoNegro) 
plt.show() 

buracoNegro = plt.imshow(imagemBuracoNegro)
```

```python
histOnibusEspacial = cv2.calcHist([imgOnibusEspacial],[0],None,[256],[0,256]) 
plt.plot(histOnibusEspacial) 
plt.show() 
onibusEspacial = plt.imshow(imagemOnibusEspacial)
```

```python
import pandas as pd
import numpy as np
import scipy
```

```python
df = pd.read_csv('Data/data.csv')
```

```python
df['Maker_Model'] = df['Make'] + " " + df['Model']
```

```python
# Select features from original dataset to form a new dataframe 
df1 = df[['Engine Fuel Type','Transmission Type','Driven_Wheels',\
          'Market Category','Vehicle Size', 'Vehicle Style',\
          'Maker_Model']]
```

```python
# For each row, combine all the columns into one column
df2 = df1.apply(lambda x: ','.join(x.astype(str)), axis=1)
```

```python
df2.head()
```

```python
# Store them in a pandas dataframe
df_clean = pd.DataFrame({'clean': df2})
```

```python
df_clean
```

```python
[print(row) for row in df_clean['clean']]
```

```python
# Create the list of list format of the custom corpus for gensim modeling 
sent = [row.split(',') for row in df_clean['clean']]
```

```python
# show the example of list of list format of the custom corpus for gensim modeling 
sent[:3]
```

```python
from gensim.models import Word2Vec
```

```python
model = Word2Vec(sent, min_count=1,size= 3,workers=3, window =3,\
                 sg = 1)
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
def display_closestwords_tsnescatterplot(model, word, size):
    
    arr = np.empty((0,size), dtype='f')
    word_labels = [word]
    close_words = model.wv.similar_by_word(word)
    arr = np.append(arr, np.array([model.wv[word]]), axis=0)
    for wrd_score in close_words:
        wrd_vector = model.wv[wrd_score[0]]
        word_labels.append(wrd_score[0])
        arr = np.append(arr, np.array([wrd_vector]), axis=0)
        
    tsne = TSNE(n_components=2, random_state=0)
    np.set_printoptions(suppress=True)
    Y = tsne.fit_transform(arr)
    x_coords = Y[:, 0]
    y_coords = Y[:, 1]
    plt.scatter(x_coords, y_coords)
    for label, x, y in zip(word_labels, x_coords, y_coords):
        plt.annotate(label, xy=(x, y), xytext=(0, 0), textcoords='offset points')
    plt.xlim(x_coords.min()+0.00005, x_coords.max()+0.00005)
    plt.ylim(y_coords.min()+0.00005, y_coords.max()+0.00005)
    plt.show()
```

```python
display_closestwords_tsnescatterplot(model, 'Porsche 718 Cayman', 3) 
```

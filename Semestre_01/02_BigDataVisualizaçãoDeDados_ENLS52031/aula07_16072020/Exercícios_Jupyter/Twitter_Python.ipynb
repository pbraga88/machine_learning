```python
import texthero as hero
import pandas as pd
```

```python
data = {'tweet': []}
df = pd.DataFrame(data=data)
```

```python
for tweet in retornoTwitter:
    df= df.append({'tweet':tweet.text}, ignore_index=True)
    
```

```python
df
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
df['tweet'] = hero.clean(df['tweet'])
```

```python
NUM_TOP_WORDS = 100
frequencia_de_palavras = hero.top_words(df['tweet'])[:NUM_TOP_WORDS]
```

```python
wordcloud = WordCloud()
wordcloud.generate_from_frequencies(frequencies=frequencia_de_palavras)
```

```python
plt.figure(figsize=(15,8))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
from texthero import stopwords
```

```python
default_stopwords = stopwords.DEFAULT

#add a list of stopwords to the stopwords
custom_stopwords = default_stopwords.union(set(["https", "co"]))
```

```python
df['tweet'] = hero.remove_stopwords(df['tweet'],\
                                          custom_stopwords)
```

```python
NUM_TOP_WORDS = 100
frequencia_de_palavras = hero.top_words(df['tweet'])[:NUM_TOP_WORDS]
```

```python
wordcloud = WordCloud()
wordcloud.generate_from_frequencies(frequencies=frequencia_de_palavras)
```

```python
plt.figure(figsize=(15,8))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

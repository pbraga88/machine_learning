# Viagens Produzidas por Modo Principal - 2017

```python
df_tipo_viagem_total = pd.DataFrame(df_tipo_viagem.iloc[:,1:15].sum(), columns=['Total']) ## 1:15 = colunas Metrô até Outros
df_tipo_viagem_total.reset_index(level=0, inplace=True)
df_tipo_viagem_total.rename(columns={"index": "Tipo"}, inplace=True)
total_viagens = df_tipo_viagem_total['Total'].sum()
df_tipo_viagem_total['Porcentagem'] = df_tipo_viagem_total['Total'].apply(lambda x: x/total_viagens * 100)
df_tipo_viagem_total.sort_values(by='Porcentagem', inplace=True, ascending=True)

total_outros = df_tipo_viagem_total.loc[df_tipo_viagem_total['Porcentagem'] < 1]['Porcentagem'].sum()
df_tipo_viagem_porcentagem = df_tipo_viagem_total.loc[df_tipo_viagem_total['Porcentagem'] >= 1][['Tipo', 'Porcentagem']]
df_tipo_viagem_porcentagem = df_tipo_viagem_porcentagem.append({'Tipo': 'Outros', 'Porcentagem': total_outros},ignore_index=True)
df_tipo_viagem_porcentagem.sort_values(by='Porcentagem', inplace=True, ascending=True)

fig, ax = plt.subplots(figsize=(12, 12), subplot_kw=dict(aspect="equal"))
patches,texts,_ = ax.pie(df_tipo_viagem_porcentagem['Porcentagem'], labels=df_tipo_viagem_porcentagem['Tipo'], autopct='%1.1f%%', startangle=90, pctdistance=.85, labeldistance=1.05,textprops={'fontsize': 14})
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

for text in texts:
  text.set_fontsize(14)

#draw circle
centre_circle = plt.Circle((0,0),0.7,fc='white')
fig.gca().add_artist(centre_circle)

ax.set_title('Viagens realizadas por modo principal - 2017', fontsize= 30)
plt.tight_layout()
plt.show()
```

```python
df_tipo_viagem_transporte = df_tipo_viagem_total.loc[df_tipo_viagem_total['Tipo'] != 'A pé'][['Tipo', 'Total']]
total_transporte = df_tipo_viagem_transporte['Total'].sum()
df_tipo_viagem_transporte['Porcentagem'] =df_tipo_viagem_transporte['Total']/total_transporte*100

total_transporte_outros = df_tipo_viagem_transporte.loc[df_tipo_viagem_transporte['Porcentagem'] < 1.3]['Porcentagem'].sum()
df_tipo_viagem_transporte_porcentagem = df_tipo_viagem_transporte.loc[df_tipo_viagem_transporte['Porcentagem'] >= 1.3][['Tipo', 'Porcentagem']]
df_tipo_viagem_transporte_porcentagem = df_tipo_viagem_transporte_porcentagem.append({'Tipo': 'Outros', 'Porcentagem': total_transporte_outros},ignore_index=True)
df_tipo_viagem_transporte_porcentagem.sort_values(by='Porcentagem', inplace=True, ascending=True)

fig, ax = plt.subplots(figsize=(12, 12), subplot_kw=dict(aspect="equal"))
wedges,texts,_ = ax.pie(df_tipo_viagem_transporte_porcentagem['Porcentagem'], labels=df_tipo_viagem_transporte_porcentagem['Tipo'], autopct='%1.1f%%', startangle=90, pctdistance=.85, labeldistance=1.1,textprops={'fontsize': 12})
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

for text in texts:
  text.set_fontsize(13)

#draw circle
centre_circle = plt.Circle((0,0),0.7,fc='white')
fig.gca().add_artist(centre_circle)

ax.set_title('Viagens realizadas com meios de transporte - 2017', fontsize= 30)
plt.tight_layout()
plt.show()
```

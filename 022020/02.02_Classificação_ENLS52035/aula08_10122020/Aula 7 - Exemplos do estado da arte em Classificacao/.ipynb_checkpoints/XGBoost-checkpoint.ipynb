# Gradient Boosting

* It works by sequentialling adding predictors to an ensenmble
* Each new model corrects its predecessor
* Let's see the code.

### Generate the data

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

np.random.seed(10)
X, y = make_moons(1000, noise=0.25)
plt.scatter(X[:, 0], X[:, 1], s=40, c=y, cmap=plt.cm.Spectral)
```

### Train the first DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg1 = DecisionTreeRegressor(max_depth=2)
tree_reg1.fit(X, y)
```

#### Now train a second DecisionTreeRegressor on the residual errors made by the first predictor:

```python
y2 = y - tree_reg1.predict(X)
tree_reg2 = DecisionTreeRegressor(max_depth=2)
tree_reg2.fit(X, y2)
```

### Then we train a third regressor on the residual errors made by the second predictor:

```python
y3 = y2 - tree_reg2.predict(X)
tree_reg3 = DecisionTreeRegressor(max_depth=2)
_ = tree_reg3.fit(X, y3)
```

```python
y_pred = sum(tree.predict(X[:10]) for tree in (tree_reg1, tree_reg2, tree_reg3))
print(y_pred)
print(y[:10])
```

# Gradient Boosting from Sklearn

```python
from sklearn.ensemble import GradientBoostingRegressor
gbrt = GradientBoostingRegressor(max_depth=2, n_estimators=3, learning_rate=1.0)
gbrt.fit(X, y)
gbrt.predict(X[:10]), y_pred
```

# Elements of Supervised Learning
* Model: how to make prediction $\hat{y}_{i}=f\left(x_{i}\right)$
    * Linear model: $\hat{y}_{i}=\sum_{j} w_{j} x_{i j}$
* Parameters: the things we need to learn from data
    * Linear model: $\Theta=\left\{w_{j} | j=1, \cdots, d\right\}$
* Objective Function: $\quad Obj(\Theta)=L(\Theta)+\Omega(\Theta)$
    * Training Loss $L(\Theta)$ measures how well model fit on training data
    * Regularization $\Omega(\Theta)$ measures complexity of model

* Linear Model: $L(\Theta)=\sum_{i}\left(\hat{y}_{i}-y_{i}\right)^{2}, \quad \Omega(\Theta)=\lambda\|w\|_{2}^{2}$

# Elements of Tree Learning
* Model: assuming we have $\mathrm{K}$ trees
    $$\hat{y}_{i}=\sum_{k=1}^{K} f_{k}\left(x_{i}\right), \quad f_{k} \in \mathcal{F}$$
* $\mathcal{F}$ corresponds to Space of Regression trees
    
    
* Objective is given by, where $l$ is the loss and $\Omega$ is the regularization (complexity)
    $$Obj=\sum_{i=1}^{n} l\left(y_{i}, \hat{y}_{i}\right)+\sum_{k=1}^{K} \Omega\left(f_{k}\right)$$

# Trade off in learning algorithms

$$Obj=\sum_{i=1}^{n} l\left(y_{i}, \hat{y}_{i}\right)+\sum_{k=1}^{K} \Omega\left(f_{k}\right)$$

* Optimizing training loss encourages predictive models
    * Fitting well in training data at least get you close to training data which is hopefully close to the underlying distribution
* Optimizing regularization encourages simple models
    * Simpler models tends to have smaller variance in future predictions, making prediction stable

# Learning the tree ensemble

* Mathematically, we can write our model in the form
$$\hat{y}_{i}=\sum_{k=1}^{K} f_{k}\left(x_{i}\right), f_{k} \in \mathcal{F}$$

* where $K$ is the number of trees, $f$ is a function in the functional space $\mathcal{F},$ and $\mathcal{F}$ is the set of all possible CARTs. The objective function to be optimized is given by

$$\operatorname{obj}(\theta)=\sum_{i}^{n} l\left(y_{i}, \hat{y}_{i}\right)+\sum_{k=1}^{K} \Omega\left(f_{k}\right)$$

# Learning the tree ensemble
* Learning tree structure is much harder than traditional optimization problem where you can simply take the gradient. 
* It is intractable to learn all the trees at once. 
* Instead, we use an additive strategy: fix what we have learned, and add one new tree at a time. 

\begin{aligned}
&\hat{y}_{i}^{(0)}=0\\
&\hat{y}_{i}^{(1)}=f_{1}\left(x_{i}\right)=\hat{y}_{i}^{(0)}+f_{1}\left(x_{i}\right)\\
&\hat{y}_{i}^{(2)}=f_{1}\left(x_{i}\right)+f_{2}\left(x_{i}\right)=\hat{y}_{i}^{(1)}+f_{2}\left(x_{i}\right)\\
&\ldots \\
&\hat{y}_{i}^{(t)}=\sum_{k=1}^{t} f_{k}\left(x_{i}\right)=\hat{y}_{i}^{(t-1)}+f_{t}\left(x_{i}\right)
\end{aligned}

# Taylor Expansion of Loss

* Goal $Obj^{(t)}=\sum_{i=1}^{n} l\left(y_{i}, \hat{y}_{i}^{(t-1)}+f_{t}\left(x_{i}\right)\right)+\Omega\left(f_{t}\right)+\text {constant }$
* Taylor expansion
    * Recall $f(x+\Delta x) \simeq f(x)+f^{\prime}(x) \Delta x+\frac{1}{2} f^{\prime \prime}(x) \Delta x^{2}$
    * Define $g_{i}=\partial_{\hat{y}^{(t-1)}} l\left(y_{i}, \hat{y}^{(t-1)}\right), \quad h_{i}=\partial_{\hat{y}^{(t-1)}}^{2} l\left(y_{i}, \hat{y}^{(t-1)}\right)$

$$O b j^{(t)} \simeq \sum_{i=1}^{n}\left[l\left(y_{i}, \hat{y}_{i}^{(t-1)}\right)+g_{i} f_{t}\left(x_{i}\right)+\frac{1}{2} h_{i} f_{t}^{2}\left(x_{i}\right)\right]+\Omega\left(f_{t}\right)+\text { constant }$$
* Example for square loss
$$g_{i}=\partial_{\hat{y}(t-1)}\left(\hat{y}^{(t-1)}-y_{i}\right)^{2}=2\left(\hat{y}^{(t-1)}-y_{i}\right) h_{i}=\partial_{\hat{y}^{(t-1)}}^{2}\left(y_{i}-\hat{y}^{(t-1)}\right)^{2}=2$$

# After we remove all the constants, the specific objective at step $𝑡$ becomes


$$
\begin{aligned}
\mathrm{obj}^{(t)} & \approx \sum_{i=1}^{n}\left[g_{i} f_{t}\left(x_{i}\right)+\frac{1}{2} h_{i} f_{t}^{2}\left(x_{i}\right)\right]+\Omega\left(f_{t}\right)\\
&=\sum_{i=1}^{n}\left[g_{i} w_{q\left(x_{i}\right)}+\frac{1}{2} h_{i} w_{q\left(x_{i}\right)}^{2}\right]+\gamma T+\frac{1}{2} \lambda \sum_{j=1}^{T} w_{j}^{2} \\
&=\sum_{j=1}^{T}\left[\left(\sum_{i \in I_{j}} g_{i}\right) w_{j}+\frac{1}{2}\left(\sum_{i \in I_{j}} h_{i}+\lambda\right) w_{j}^{2}\right]+\gamma T
\end{aligned}
$$

where $\boldsymbol{I}_{j}=\left\{i | q\left(x_{i}\right)=j\right\}$ is the set of indices of data points assigned to the $𝑗$-th leaf


# Final equation, finally 😅

We could further compress the expression by defining $G_{j}=\sum_{i \in I_{j}} g_{i}$ and $H_{j}=\sum_{i \in I_{j}} h_{i}:$

$$\mathrm{obj}^{(t)}=\sum_{j=1}^{T}\left[G_{j} w_{j}+\frac{1}{2}\left(H_{j}+\lambda\right) w_{j}^{2}\right]+\gamma T$$

In this equation, $w_{j}$ are independent with respect to each other, the form $G_{j} w_{j}+\frac{1}{2}\left(H_{j}+\lambda\right) w_{j}^{2}$ is quadratic and the best $w_{j}$ for a given structure $q(x)$ and the best objective reduction we can get is:
$$
\begin{aligned}
w_{j}^{*} &=-\frac{G_{j}}{H_{j}+\lambda} \\
\operatorname{obj}^{*} &=-\frac{1}{2} \sum_{j=1}^{T} \frac{G_{j}^{2}}{H_{j}+\lambda}+\gamma T
\end{aligned}
$$

# Learn the tree structure

* Now that we have a way to measure how good a tree is, ideally we would enumerate all possible trees and pick the best one.
* In practice this is intractable, so we will try to optimize one level of the tree at a time. Specifically we try to split a leaf into two leaves, and the score it gains is

$$Gain=\frac{1}{2}\left[\frac{G_{L}^{2}}{H_{L}+\lambda}+\frac{G_{R}^{2}}{H_{R}+\lambda}-\frac{\left(G_{L}+G_{R}\right)^{2}}{H_{L}+H_{R}+\lambda}\right]-\gamma$$

1) the score on the new left leaf 2) the score on the new right leaf 3) The score on the original leaf 4) regularization on the additional leaf.

# Working on practice

```python
from sklearn import datasets
import xgboost as xgb

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

# Split data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.3, random_state=43)
```

# Setting up our data with XGBoost

```python
D_train = xgb.DMatrix(X_train, label=Y_train)
D_test = xgb.DMatrix(X_test, label=Y_test)
```

# Defining an XGBoost model


```python
# set xgboost params
param = {
    'max_depth': 3,  # the maximum depth of each tree
    'eta': 0.5,  # the training step for each iteration
    'silent': 1,  # logging mode - quiet
    'objective': 'multi:softprob',  # error evaluation for multiclass training
    'num_class': 3}  # the number of classes that exist in this datset
steps = 20  # the number of training iterations
```

```python
model = xgb.train(param, D_train, steps)
```

# Let’s now run an evaluation

```python
import numpy as np
from sklearn.metrics import precision_score, recall_score, accuracy_score

preds = model.predict(D_test)
best_preds = np.asarray([np.argmax(line) for line in preds])

print("Precision = {}".format(precision_score(Y_test, best_preds, average='macro')))
print("Recall = {}".format(recall_score(Y_test, best_preds, average='macro')))
print("Accuracy = {}".format(accuracy_score(Y_test, best_preds)))
```

# Improving Evaluation

```python
from sklearn.model_selection import GridSearchCV

clf = xgb.XGBClassifier(n_estimators=20,learning_rate=0.5,max_delta_step=5)
parameters = {
     "eta"    : [0.5, 0.10, 0.25, 0.30, 0.35, 0.4 ] ,
     "max_depth"        : [ 3, 4, 15],
     "min_child_weight" : [ 1, 3, 5, 7 ],
     "gamma"            : [ 0.0, 0.1, 0.4 ],
     "colsample_bytree" : [ 0.3, 0.4, 0.5 , 0.7 ],
     'num_class': [3],
     'objective': ['multi:softprob']
     }

grid = GridSearchCV(clf,
                    parameters, n_jobs=4,
                    scoring="neg_log_loss",
                    cv=3)

_ = grid.fit(X_train, Y_train)
```

# Test Model

```python
preds = grid.predict(X_test)

print("Precision = {}".format(precision_score(Y_test, preds, average='macro')))
print("Recall = {}".format(recall_score(Y_test, preds, average='macro')))
print("Accuracy = {}".format(accuracy_score(Y_test, preds)))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

log_clf = LogisticRegression()
rnd_clf = RandomForestClassifier()
svm_clf = SVC(probability=True)

voting_clf = VotingClassifier(
    estimators=[('lr', log_clf), ('rf', rnd_clf), ('svc', svm_clf)],
    voting='soft'
)

_ = voting_clf.fit(X_train, Y_train)
```

```python
preds = voting_clf.predict(X_test)

print("Precision = {}".format(precision_score(Y_test, preds, average='macro')))
print("Recall = {}".format(recall_score(Y_test, preds, average='macro')))
print("Accuracy = {}".format(accuracy_score(Y_test, preds)))
```

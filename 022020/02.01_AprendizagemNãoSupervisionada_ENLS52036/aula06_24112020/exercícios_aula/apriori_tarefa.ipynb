```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
import pandas as pd
```

```python
df=pd.read_csv('store_data.csv',header=None)
df.head()
```

```python
#gerar a lista de listas trans
trans=[]
for i in range(len(df)):
    trans.append([x for x in list(df.iloc[i]) if str(x) != 'nan'])
```

```python
trans
```

```python
#rodar apriori
results = list(apriori(trans, min_support=0.002))#, min_lift=1.3))
```

```python
#mostrar resultados
# association_results
results
```

```python
for i in results:
    if len(i[0]) == 4:
        print(i[1])
```

```python
maxi = 0.0
indexMaxi = 0
for i in results:
    if len(i[0]) == 4 and i[1] > maxi:
        maxi = i[1]
        indexMaxi = i
maxi, indexMaxi[0]
```

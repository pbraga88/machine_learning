```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
import pandas as pd
import random
```

```python
df=pd.read_csv('store_data.csv',header=None)
df.head()
```

```python
len(df)
```

```python
#gerar a lista de listas trans
trans=[]
for i in range(len(df)):
  trans.append([x for x in list(df.iloc[i]) if str(x) != 'nan'])

```

```python
#rodar apriori
results=list(apriori(trans,min_support=0.002))
```

```python
trans
```

```python
#mostrar resultados (controla as combinações mostradas 1,2,3...)
quant=4
for r in results:
  if len(r[0])==quant:
    print(r)
```

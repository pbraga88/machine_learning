```python
import numpy as np
```

```python
X=np.array(dfn)
```

```python
X.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit(X)
XS=scaler.transform(X)
```

```python
XS[:,190].mean(),XS[:,190].std()
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=5,random_state=5)
km.fit(XS)
km.inertia_
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
lk=[]
lc=[]
ls=[]
for i in range(2,21):
 km=KMeans(n_clusters=i,random_state=51)
 km.fit(XS)
 lk.append(i)
 lc.append(km.inertia_)
 ls.append(silhouette_score(X, km.labels_))
plt.plot(lk,lc,color='black') 
plt.show()
plt.plot(lk,ls,color='blue') 
plt.show()
```

```python
ls#...lista de silhuettes em ordem decrescente...3 parece ser a opção mesno ruim
```

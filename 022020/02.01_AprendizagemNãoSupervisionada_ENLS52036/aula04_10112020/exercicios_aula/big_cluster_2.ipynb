```python
import numpy as np
```

```python
X=np.array(dfn) # Transforma dfn em array, visto que são todas linhas numéricas
X.shape
```

```python
from sklearn.preprocessing import StandardScaler # import para fazer standardização das features
scaler=StandardScaler()
```

```python
scaler.fit(X)
XS=scaler.transform(X) # XS = X Standardizado
```

```python
# Verificando a standardização variáveis aleatórias
for i in range(150, 155):
    print("Média da coluna", i, ":", XS[:,i].mean())
    print("Desvio Padrão da coluna", i, ":", (XS[:,i].std()))
```

```python
from sklearn.cluster import KMeans
```

```python
# Roda o kmeans para ver o erro:
# Somatório da distância ao quadrado de cada ponto à centróride de seu cluster
km=KMeans(n_clusters=5,random_state=5)
km.fit(XS)
km.inertia_
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
lk=[] # Aqui irá a quantidade clusters (varíavel i do laço for abaixo) 
lc=[] # Nesta lista vai o custo (lm_inertia)
ls=[] # E nesta o silhouete
for i in range(2,21): # de 2 até 21 clusters
    km=KMeans(n_clusters=i,random_state=51)
    km.fit(XS)
    lk.append(i)
    lc.append(km.inertia_)
    ls.append(silhouette_score(X, km.labels_))
plt.plot(lk,lc,color='black')    
plt.show()
plt.plot(lk,ls,color='blue')  
plt.show()
```

Verificando qual cidade em cada cluster mais se aproxima da centróide. Em outras palavras, qual cidade melhor representa cada um dos clusters

```python
ld0,ld1,ld2=[],[],[]

# Percorre as features standardizadas XS
for i in range(len(XS)):
    if km.labels_[i] == 0: # Se, em cluster 0
        ld0.append((i, np.linalg.norm(XS[i] - km.cluster_centers_[0]))) # Adiciona à lista ld0, o ponto 'i'
                                                 # mais a distância daquele ponto para a centróide do cluster 0
    elif km.labels_[i] == 1: # Se, em cluster 1
        ld1.append((i, np.linalg.norm(XS[i] - km.cluster_centers_[1])))# Adiciona à lista ld1, o ponto 'i'
                                                 # mais a distância daquele ponto para a centróide do cluster 1
    elif km.labels_[i] == 2: # Por fim, se, em cluster 2
        ld2.append((i, np.linalg.norm(XS[i] - km.cluster_centers_[2])))# Adiciona à lista ld2, o ponto 'i'
                                                 # mais a distância daquele ponto para a centróide do cluster 2
```

```python
# Organizando as listas (sort) pela distância de cada ponto à centróide com o auxílio da função lambda
ld0.sort(key = lambda x:x[1]) # sort pelo segundo elemento da lista
ld1.sort(key = lambda x:x[1])
ld2.sort(key = lambda x:x[1])

# Exemplo do sort com o cluster 2
ld2
```

```python
# As cidades que melhor representam cada um dos clusters
print("Cluster 0:", mun[ld0[0][0]])
print("Cluster 1:",mun[ld1[0][0]])
print("Cluster 2:",mun[ld2[1][0]])
```

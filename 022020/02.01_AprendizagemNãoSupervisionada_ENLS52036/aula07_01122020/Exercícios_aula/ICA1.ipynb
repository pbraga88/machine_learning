```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random

from sklearn.decomposition import FastICA


S=[]
d=math.pi*0.1
for i in range(30000):
    S.append([math.sin(i*d)+random.random(),np.sign(math.sin((i+2)*d))])
S=np.array(S)
A = np.array([[2, 1], [1, 2]]) 
X = np.dot(S, A.T)
ica = FastICA(random_state=42)
S_ica_ = ica.fit(X).transform(X) 


plt.plot(X[:50,0])
plt.plot(X[:50,1])
plt.title('Combinados')

plt.show()

plt.plot(S[:50,0])
plt.plot(S[:50,1])
plt.title('Separados originais')

plt.show()

plt.plot(S_ica_[:50,0])
plt.plot(S_ica_[:50,1])
plt.title('Separados ICA')


plt.show()
```
